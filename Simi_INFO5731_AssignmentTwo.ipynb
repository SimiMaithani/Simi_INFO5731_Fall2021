### (2) Basic Text Pre-processing of text data







```python
import pandas as pd
df = pd.read_csv("cases.txt", sep="\n")
df.rename(columns = {'5 Ala. 740':'cases'}, inplace = True)
df

# Part (2) below, in one cell to reduce the number of cells for the code
# Lower case
df['cases'] = df['cases'].apply(lambda x: " ".join(x.lower() for x in x.split()))

# Removing Punctuations
import string
df['cases'] = df['cases'].str.replace('[{}]'.format(string.punctuation), '')

# Removing Stopwords
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
stop=stopwords.words('english')
df['cases'] = df['cases'].apply(lambda x:" ".join(x for x in x.split() if x not in stop))

# Frequent word removal
freq=pd.Series(' '.join(df['cases']).split()).value_counts()[:10]
freq=list(freq.index)
df['cases'] = df['cases'].apply(lambda x:" ".join(x for x in x.split() if x not in freq))

# Rare word removal
freq=pd.Series(' '.join(df['cases']).split()).value_counts()[-10:]
freq=list(freq.index)
df['cases'] = df['cases'].apply(lambda x:" ".join(x for x in x.split() if x not in freq))

# Spelling Correction
from textblob import TextBlob
df['cases'] = df['cases'].apply(lambda x: str(TextBlob(x).correct()))

# Stemming
from nltk.stem import PorterStemmer
st=PorterStemmer()
df['cases'] = df['cases'].apply(lambda x: " ".join([st.stem(word) for word in x.split()]))

# Lemmatization
from textblob import Word
nltk.download('wordnet')
df['cases'] = df['cases'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))

# Tokenization
import nltk
nltk.download('punkt')
df['cases'] = df['cases'].apply(lambda x: TextBlob(x).words)

# (3) Saving to a csv file
df.to_csv('processed_file.csv')
print(df)
```
