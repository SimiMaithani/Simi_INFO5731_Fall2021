```python
# Write your code here
# saving the imported file in a dataframe 
import pandas as pd
df = pd.read_csv('citeseer1_file.csv')
df
```

```python
# Importing the important libraries
import urllib.request
import re
import requests
import pandas as pd

# Performing basic pre-processing to remove inconsistencies from the data

# Converting to lower case
df['Abstract']= df['Abstract'].astype(str)
df['Abstract'] = df['Abstract'].apply(lambda x: " ".join(x.lower() for x in x.split()))

# Removing Punctuations
import string
df['Abstract'] = df['Abstract'].str.replace('[{}]'.format(string.punctuation), '')

# Removing stopwords
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
stop=stopwords.words('english')
df['Abstract'] = df['Abstract'].apply(lambda x:" ".join(x for x in x.split() if x not in stop))

# common words removal
freq=pd.Series(' '.join(df['Abstract']).split()).value_counts()[:11]
freq=list(freq.index)
df['Abstract'] = df['Abstract'].apply(lambda x:" ".join(x for x in x.split() if x not in freq))

# Rare words removal
freq=pd.Series(' '.join(df['Abstract']).split()).value_counts()[-10:]
freq=list(freq.index)
df['Abstract'] = df['Abstract'].apply(lambda x:" ".join(x for x in x.split() if x not in freq))
pd.set_option("display.max_colwidth", None)
df.head()
```

```python
# Spelling correction
from textblob import TextBlob
df['Abstract'] = df['Abstract'].apply(lambda x: str(TextBlob(x).correct()))
df.head()
```

```python
import urllib.request
import re
import requests
import pandas as pd
from nltk.stem import PorterStemmer
st=PorterStemmer()

# Changing the data type of the Abstract column to string
df['Abstract']= df['Abstract'].astype(str)
# Stemming
from nltk.stem import PorterStemmer
st=PorterStemmer()
df['Abstract'] = df['Abstract'].apply(lambda x: " ".join([st.stem(word) for word in x.split()]))

# Lemmatization
from textblob import Word
nltk.download('wordnet')
df['Abstract'] = df['Abstract'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))

# Tokenization
from textblob import TextBlob
import nltk
nltk.download('punkt')
df['Abstract'] = df['Abstract'].apply(lambda x: TextBlob(x).words)

df.head()
# Saving the clean file into another csv
df.to_csv('processed_citeseer2.csv')
df.head()
```

```python
import collections
from nltk import ngrams
from nltk.tokenize import word_tokenize
import itertools

# Counting the frequency of all the n-grams
freq_list = []
for sentence in df['Abstract']:
  freq_list.append(sentence)   
condition = [x for x in freq_list if x != []]
iterations = list(itertools.chain.from_iterable(condition))
# Frequency for all the trigrams

tri_grams = nltk.trigrams(iterations)
frequency_dist = nltk.FreqDist(tri_grams)
tri_grams_dict = dict(frequency_dist)
for word in tri_grams_dict:
    print( str(word) + ': '+ str(frequency_dist[word]))
```
