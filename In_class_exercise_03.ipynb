```python
# The feature selection method I will be using is the Term Frequency-Inverse Document Frequency. This method comes 
# under the Filter model. The most important features are stopword removal and tokenization where tokenization holds more importance than 
# stopwords removal since cutting the sentence into words is the most step amonngst all.
# The descending order in which the feature extraction methods should be placed are: 
# Tokenization -> stop words removal -> special characters removal -> number of lower cases -> numeric removal

tf1 = (dataset['review'][1:2]).apply(lambda x: pd.value_counts(x.split(" "))).sum(axis = 0).reset_index()
# A new column called 'tf' for storing the frequency value
tf1.columns = ['words','tf']
tf1 

# Inverse Document Frequency
import numpy as np

for i,word in enumerate(tf1['words']):
  tf1.loc[i, 'idf'] = np.log(dataset.shape[0]/(len(dataset[dataset['review'].str.contains(word)])))

tf1
```

```python
# TF-IDF is the multiplication of the TF and IDF which we calculated above.
# We can see that the TF-IDF has penalized words like ‘well’ and ‘real' because they are 
# commonly occurring words. However, it has given a high weight to “realism” since that will be very useful in determining the sentiment of the tweet.
tf1['tfidf'] = tf1['tf'] * tf1['idf']
tf1
```
